```python
import pandas as pd
import requests
from io import BytesIO

def load_excel_from_url(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Check if the request was successful
        # Load Excel file into a DataFrame
        df = pd.read_excel(BytesIO(response.content))
        print("Excel file loaded successfully!")
        return df
    except requests.exceptions.HTTPError as e:
        print(f"HTTP error: {e}")
    except requests.exceptions.RequestException as e:
        print(f"Error downloading the file: {e}")

def load_excel_from_local(file_path):
    try:
        df = pd.read_excel(file_path)
        print("Excel file loaded successfully!")
        return df
    except FileNotFoundError:
        print("Local file not found. Please check the path.")

# Choose either method
url = 'https://drive.google.com/uc?export=download&id=YOUR_FILE_ID'  # Replace with direct link
file_path = r"C:\Users\deeps\Downloads\Output Data Structure.xlsx"   # Replace with your actual file path

# Load from URL or local based on availability
df = load_excel_from_url(url) or load_excel_from_local(file_path)

# If DataFrame loaded successfully, check for 'Link' column
if df is not None:
    print("Columns in the DataFrame:", df.columns)
    if 'Link' in df.columns:
        for index, row in df.iterrows():
            link = row['Link']
            print(link)
    else:
        print("Column 'Link' not found in the DataFrame.")
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Define headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36"
}

# Load the DataFrame from an Excel file
try:
    df = pd.read_excel(r"C:\Users\deeps\Downloads\Output Data Structure.xlsx")
except Exception as e:
    print("Error reading Excel file:", e)

# Load positive and negative words lists
try:
    with open(r"C:\Users\deeps\Downloads\positive-words.txt") as post_file:
        positive_words = {line.strip().lower() for line in post_file}

    with open(r"C:\Users\deeps\Downloads\negative-words.txt") as neg_file:
        negative_words = {line.strip().lower() for line in neg_file}
except Exception as e:
    print("Error reading positive or negative words file:", e)

# Function to filter out non-visible elements in HTML
def tag_visible(element):
    return element.parent.name not in ['style', 'script', 'head', 'title', 'meta', '[document]']

# Function to extract visible text from HTML content
def text_from_html(html_content):
    soup = BeautifulSoup(html_content, "html.parser")
    texts = soup.findAll(string=True)
    visible_texts = filter(tag_visible, texts)
    return " ".join(t.strip() for t in visible_texts)

# Process each URL in the DataFrame
for i in range(len(df)):
    try:
        url = df.iloc[i, 1]  # Adjust if URL column isn't the second column
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Check if request was successful

        # Extract text from HTML
        blog_paragraph = text_from_html(response.text)

        # Initialize counts
        positive_count = 0
        negative_count = 0

        # Split text into words and count occurrences (case insensitive)
        words = blog_paragraph.lower().split()
        for word in words:
            if word in positive_words:
                positive_count += 1
            elif word in negative_words:
                negative_count += 1

        # Store the counts in the DataFrame
        df.at[i, 'Positive Score'] = positive_count
        df.at[i, 'Negative Score'] = negative_count

    except requests.exceptions.RequestException as req_err:
        print(f"Request error for URL {url}: {req_err}")
        df.at[i, 'Positive Score'] = None
        df.at[i, 'Negative Score'] = None
    except Exception as e:
        print(f"Error processing URL {url}: {e}")

# Save the DataFrame to an Excel file
try:
    df.to_excel(r"C:\Users\deeps\Downloads\Output_Data_Structure.xlsx", index=False)
    print("Excel file saved successfully.")
except Exception as e:
    print("Error saving Excel file:", e)
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from bs4.element import Comment

# Define headers for the request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36"
}

# Function to filter visible text from HTML
def tag_visible(element):
    return not (element.parent.name in ['style', 'script', 'head', 'title', 'meta', '[document]'] or isinstance(element, Comment))

# Function to extract visible text from HTML
def text_from_html(body):
    soup = BeautifulSoup(body, 'html.parser')
    texts = soup.findAll(text=True)
    visible_texts = filter(tag_visible, texts)
    return u" ".join(t.strip() for t in visible_texts)

# Load positive and negative word lists
with open(r"C:\Users\deeps\Downloads\positive-words.txt") as pos_file:
    positive_words = {line.strip().lower() for line in pos_file}

with open(r"C:\Users\deeps\Downloads\negative-words.txt") as neg_file:
    negative_words = {line.strip().lower() for line in neg_file}

# URL to analyze
url = 'https://insights.blackcoffer.com/rise-of-telemedicine-and-its-impact-on-livelihood-by-2040-3-2/'

# Send request to the webpage
response = requests.get(url, headers=headers)
blog_paragraph = text_from_html(response.text)

# Split the blog text into words and count positive/negative occurrences
positive_count = sum(1 for word in blog_paragraph.lower().split() if word in positive_words)
negative_count = sum(1 for word in blog_paragraph.lower().split() if word in negative_words)

# Calculate scores
total_words = len(blog_paragraph.split())
total_sentences = blog_paragraph.count('.') + blog_paragraph.count('!') + blog_paragraph.count('?')

Polarity_Score = (positive_count - negative_count) / ((positive_count + negative_count) + 0.000001)
Subjectivity_Score = (positive_count + negative_count) / (total_words + 0.000001)
Average_Sentence_Length = total_words / (total_sentences if total_sentences > 0 else 1)

# Display scores
print("Positive Count:", positive_count)
print("Negative Count:", negative_count)
print("Polarity Score:", Polarity_Score)
print("Subjectivity Score:", Subjectivity_Score)
print("Average Sentence Length:", Average_Sentence_Length)

# Load existing DataFrame or create a new one
try:
    df = pd.read_excel(r"C:\Users\deeps\Downloads\Output Data Structure.xlsx")
except FileNotFoundError:
    df = pd.DataFrame(columns=["URL", "Positive Count", "Negative Count", "Polarity Score", "Subjectivity Score", "Average Sentence Length"])

# Create a new DataFrame for the current row
new_row = pd.DataFrame({
    "URL": [url],
    "Positive Count": [positive_count],
    "Negative Count": [negative_count],
    "Polarity Score": [Polarity_Score],
    "Subjectivity Score": [Subjectivity_Score],
    "Average Sentence Length": [Average_Sentence_Length]
})

# Append the new row to the existing DataFrame
df = pd.concat([df, new_row], ignore_index=True)

# Save to Excel
df.to_excel(r"C:\Users\deeps\Downloads\Output Data Structure.xlsx", index=False)
print("Data saved to Excel file.")
```
